```python
import gradio as gr
def add(first,second):
    return first+second
demo1 =gr.Interface(
    fn=add,
    inputs=['number','number'],
    outputs='number',
    title='cal',
    description=f'{first}+{second}'
)
demo1.launch(
    inbrowser=True,
    inline=False,
    debug=True,
    share=True
)
```

```python
demo1.close()
```
